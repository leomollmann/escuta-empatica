```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import numpy as np

x = np.arange(len(control['pessoa A']['avg']) * 2)

legends = [Line2D([0], [0], color='b', label='Tendência dos dados extendidos'),
           Line2D([0], [0], color='g', label='Tendência dos dados por exemplos'),
           Line2D([0], [0], color='r', label='Tendência dos dados de controle')
]

def interleave_lists(a, b):
    return [val for pair in zip(a, b) for val in pair]

plt.plot(x, interleave_lists(control['pessoa A']['avg'], control['pessoa B']['avg']), 'r')
plt.plot(x, interleave_lists(extended['agente intrigante']['avg'], extended['agente empático']['avg']), 'b')
plt.plot(x, interleave_lists(example['agente intrigante']['avg'], example['agente empático']['avg']), 'g')

plt.yticks(np.arange(-1, 1.1, 0.2))

plt.xlabel('Número da interação')
plt.ylabel('Energia')
plt.title(f'Tendência A x B')
plt.legend(handles=legends)
plt.show()
```
