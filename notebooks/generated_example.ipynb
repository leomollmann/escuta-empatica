```python
import os
from ee.openai import prompt
import random
import tiktoken

examples = []

for i in range(1,24):
    with open(f'../dados/juliano_rigatti_en_extended/{i}.txt', 'r') as file:
        examples.append(file.read())

examples

def get_random_example(num_examples):
    return random.sample(examples, num_examples)

def generate_files(destination_folder, num_files):
    # Create the destination folder if it doesn't exist
    os.makedirs(destination_folder, exist_ok=True)
    
    # Generate the specified number of files
    for i in range(1, num_files + 1):
        file_path = os.path.join(destination_folder, f"{i}.txt")

        examples = get_random_example(6)
        
        # Open the file in write mode
        with open(file_path, 'w') as f:
            # Write the content to the file
            payload = f"""Create a brand new conversation flow like these examples, with at least 12 interactions. The goal is to simulate empathetic listening.
Use these as examples:
1. 
{examples[0]}

2. 
{examples[1]}

3. 
{examples[2]}

4. 
{examples[3]}

5. 
{examples[4]}

6. 
{examples[5]}
                             
7.
"""
            
            encoding = tiktoken.encoding_for_model('gpt-3.5-turbo')
            print(len(encoding.encode(payload)))

            content = prompt(payload)
            

            f.write(content)

# Example usage
destination_folder = "../dados/generated_example"
num_files = 30
generate_files(destination_folder, num_files)
```
